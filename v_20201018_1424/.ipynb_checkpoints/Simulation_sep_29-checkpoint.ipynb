```python
class ridt_random:
    
    def random_dates_list(start, end, n):
        """ recieves a START date
        and an END date
        and a N number of random Dates"""
        start_u = start.value//10**9
        end_u = end.value//10**9
        return pd.DatetimeIndex((10**9*np.random.randint(start_u, end_u, n, dtype=np.int64)).view('M8[ns]')) 
```

```python
class ridt_plot:
    def plot_age_histogram(a_ds,wid=5,title_='Normal Distribution Histogram'):
        """for a given seires this function plots a histogram
        a_ds is the series
        wid=5 is the Width (can be changed)
        title_ is the title
        """
        hist,bin_edges = np.histogram(a_ds)
        plt.figure(figsize=[8,5])
        plt.bar(bin_edges[:-1], hist, width = wid, color='#0504aa',alpha=0.7)
        plt.xlim(min(bin_edges), max(bin_edges))
        plt.grid(axis='y', alpha=0.75)
        plt.xlabel('Value',fontsize=15)
        plt.ylabel('Frequency',fontsize=15)
        plt.xticks(fontsize=15)
        plt.yticks(fontsize=15)
        plt.ylabel('Frequency',fontsize=15)
        plt.title(title_,fontsize=15)
        plt.show()
        
    def plot_bar(ds,col_data,col_x_axis,x_title,y_title,title):
        """ds is the dataset
        col_data is the column with the data
        col_x_axis is the column with the x axis
        x_title and y_title and title are the titles"""
        if col_x_axis=='index':
            x_axis_ = ds.index
        else:
            x_axis_ = ds['col_x_axis']
        y_pos = np.arange(len(x_axis_))
        plt.bar(y_pos, ds[col_data], align='center', alpha=0.5)
        plt.xticks(y_pos,x_axis_)
        plt.ylabel(y_title)
        plt.xlabel(x_title)
        plt.title(title)

        plt.show()
```

```python
class ridt_demographic:
    #population demographic distribution global
    pop_age_dist={'25-35':0.30,
              '36-45':0.16,
              '46-55':0.14,
              '56-60':0.13,
              '61-65':0.07,
              '66-75':0.11,
              '76-85':0.06,
              '90+':0.03}
    pop_age_mid={'25-35':30,
              '36-45':40,
              '46-55':50,
              '56-60':58,
              '61-65':63,
              '66-75':70,
              '76-85':80,
              '90+':90}
    pop_age_dist_ds = pd.DataFrame.from_dict(pop_age_dist,orient='index',columns=['distribution'])
    pop_age_elderlies_dist={'66-75':0.6,'76-85':0.3,'90+':0.1}
    pop_age_elderlies_mid= {'66-75':70,'76-85':80,'90+':90}   
    age_group_var = {22:2,30:5,40:5,50:5,58:2,63:2,70:5,80:5,90:5}
    
    def death_date_calculator2(ag,a = 0.0009,b = 0.3313,c=62):
        """for a given AGENT and a A B C parameters
            The Function calculates the Manhattan death Age based on the birthdate
            a one deviation random is calcualted and multipled by 10 (for ten years)
            and that is the offset of the Age
            if the Death Age is less than the entrance age
            than a new random age is given that is the entrance age with 10-15 years offset
        """
        #explanation to Journal + used 100 years statistics + trial and error calibration
        #checked the average age of people who died at certain eyars
        bd = ag.birth_date
        ed = ag['start_action']
        y_n = bd.year-1900 #Year normal
        mdy = int(np.power(y_n,2) * a + b*y_n+c) # mdy manhattan death year
        death_date_ = pd.to_datetime(str(bd.year+mdy)+"-1-1")
        random_number_of_years = int(np.random.normal(scale=1)*10) #+- 10 years
        death_date_ = death_date_ +  pd.DateOffset(years=random_number_of_years)
        if (death_date_)<ed:
            death_date_ =  ed +  pd.DateOffset(years=(random.randint(10,15)))
        return death_date_
    
    def setMaxDeathAge(DeathDate):
        """If Given Date is beyond 2099-12-31 the simulation will set it for 2101 to end the simulation"""
        if (DeathDate > pd.to_datetime('2099-12-31')):
            return pd.to_datetime('2101-1-1')
        else:
            return DeathDate
    
    def age_group_fun(num):
        return ridt_demographic.age_group_var[num]
    
    def sample_ages(pop_size,choice='other', ag = list(pop_age_mid.values()),p_ = list(pop_age_dist.values())):
        """The Function Returns a sample of Ages according to the given distribtion (With slight variation)
        pop_size: is the size of the poulation (for example pop_size=1000)
        ag: is the age group (for example elderlis 75,85,90)
        p_: is the actual distribution of the group
        """
        if  choice=='Young':
            ag = list(ridt_demographic.pop_age_mid.values())
            p_ = list(ridt_demographic.pop_age_dist.values())
        elif choice=='Old':
            ag = list(ridt_demographic.pop_age_elderlies_mid.values())
            p_ = list(ridt_demographic.pop_age_elderlies_dist.values()) 
        sam = np.random.choice(ag, pop_size, p=p_)
        #Add VAriance
        sam_variance = np.array(list(map(ridt_demographic.age_group_fun, sam)))*np.random.normal(scale=0.37,size=pop_size)
        sam2 = sam + sam_variance
        return sam2.astype(int)
```

```python
class ridt_agent:
    def get_building_info_from_agent(agent):
        """For a given AGENT returns the building data from the JSON FIle"""
        bld_key = agent['bld_key']
        building_info = json_bldgs.loc[json_bldgs['bld_key']==bld_key]
        return building_info

    def helper_function_2_cal_income(a_type,cat):
        """Returns minimum and maximum income"""
        ami_min = ridt_info.affordable_models[cat][a_type]['ami_percent_min']
        ami_max = ridt_info.affordable_models[cat][a_type]['ami_percent_max']
        income_min = int(ami_min*ridt_info.ami_100_percent)
        income_max = int(ami_max*ridt_info.ami_100_percent)
        return ([income_min,income_max])
    
    def helper_function_1_cal_income(a_type,cat):
        """returns a random income based on range and house type that is calculated"""
        income_min,income_max = ridt_agent.helper_function_2_cal_income(a_type,cat)
        income = random.randint(income_min, income_max)
        return income
    
    def get_agents_new_income_NonWIRE(agent):
        """returns a random income based on range and house type that is calculated"""
        bi = get_building_info_from_agent(agent) # get data of the building that the agent lived in
        cb = bi.squeeze() #convert to simple series
        type_ = ridt_info.affordable_models[cb['affordable_Cat']]['type']
        cat = cb['affordable_Cat']
        if (type_=='default'):
            # default is the standard rule for building ffordablity
            a_type = cb['affordable_sub']
            income = ridt_agent.helper_function_1_cal_income(a_type,cat)
        else:
            #Exception some agents will have the a income and others will have the b income
            chance = random.randint(0,10)
            if chance<6:
                income = ridt_agent.helper_function_1_cal_income('a',cat)
            else:
                income = ridt_agent.helper_function_1_cal_income('b',cat)
        return income
```

```python
class ridt_bldgs:
    def generate_properties_in_bldg(cb):
        """The Function recieves the current building 
        and returns the properties and it's properties"""
        total_units = cb['Total Unit'] #Total number of Agents in the building
        bld_key = [cb['bldgs_id']]*total_units #Defines the number of the building
        property_ds = {'app_index' : [i for i in range(total_units)],
                      'bld_key' : bld_key} # define the dictionary for the dataset
        App_datasets = pd.DataFrame.from_dict(property_ds)
        #The building key and the door number
        App_datasets['address_door']=App_datasets['bld_key']+"_"+App_datasets['app_index'].astype(str)
        #Building Name
        App_datasets['Building Name'] = cb['Bldg Proje']
        #Building construction year
        construct_year = pd.to_datetime(cb['cnstrct_yr'].astype(str)+"-1-"+"1")
        #Construction year
        App_datasets['built_year']= construct_year
        #is the apartment orginaly affordable?
        App_datasets['affordable_living'] = 0
        # by default all the apartments are not for elderly
        App_datasets['elderly_lving'] = 0
        #by default apartments are non rnetal
        App_datasets['Rental_Units'] = 0
        #by default apartments are non rental
        App_datasets['ApartmentType'] = 'Other'
        if cb['Rental_Units']>0:
            App_datasets.loc[0:cb['Rental_Units']-1,'Rental_Units'] = 1
            App_datasets.loc[0:cb['Rental_Units']-1,'ApartmentType'] = 'Rental'
        App_datasets['Ap_sq_feet_size'] = 1364 # default
        App_datasets['Ap_number_of_rooms'] = 2 # default
        if cb['protected']=='1':
            App_datasets['protected']=True
        else:
            App_datasets['protected']=False
        ela = cb['elderly_affordable'] #elderly living agents
        if ela>0:
            # Check if it's wire if not all other cases
            App_datasets.loc[0:ela,'elderly_lving'] = 1
        if cb['Afford_uni']>0:
            App_datasets.loc[0:cb['Afford_uni'],'affordable_living'] = 1
        return App_datasets
    
    def afford_price_range_NotWIRE(cb,App_datasets):
        """generates income range min and max for Non WIRE
        in addition it generates Upper Income
        cb: current building
        App_datasets: property dataset in the current building"""
         #calculate affordability according to the building
        if cb['Afford_uni']>0:
            #Needs modification
            type_ = ridt_info.affordable_models[cb['affordable_Cat']]['type']
            cat = cb['affordable_Cat']
            if (type_=='default'):
                a_type = cb['affordable_sub']
                income_min,income_max = ridt_agent.helper_function_2_cal_income(a_type,cat)
                App_datasets.loc[0:cb['Afford_uni']-1,'Low_income'] = income_min
                App_datasets.loc[0:cb['Afford_uni']-1,'Upper_income'] = income_max
                #Upper Income Needs 
                App_datasets.loc[cb['Afford_uni'] :,'Low_income'] = ridt_info.non_affordable['min']
                App_datasets.loc[cb['Afford_uni'] :,'Upper_income'] = ridt_info.non_affordable['max']

            else:
                a_ap_percent = ridt_info.affordable_models[cb['affordable_Cat']]['a']['ap_percent']
                a_max = int(np.floor(a_ap_percent*cb['Afford_uni']))
                income_min,income_max = ridt_agent.helper_function_2_cal_income('a',cat)
                App_datasets.loc[0:a_max-1,'Low_income'] = income_min
                App_datasets.loc[0:a_max-1,'Upper_income'] = income_max
                b_ap_percent = ridt_info.affordable_models[cb['affordable_Cat']]['b']['ap_percent']
                b_max = a_max+int(np.floor(b_ap_percent*cb['Afford_uni']))
                income_min,income_max = ridt_agent.helper_function_2_cal_income('b',cat)
                App_datasets.loc[a_max:a_max+b_max-1,'Low_income'] = income_min
                App_datasets.loc[a_max:a_max+b_max-1,'Upper_income'] = income_max
                #upper income
                App_datasets.loc[a_max+b_max:,'Low_income'] = ridt_info.non_affordable['min']
                App_datasets.loc[a_max+b_max:,'Upper_income'] = ridt_info.non_affordable['max']
        else:
            """for a given building the function claculates the income"""
            App_datasets['Low_income'] = ridt_info.non_affordable['min']
            App_datasets['Upper_income'] = ridt_info.non_affordable['max']
        return App_datasets
 
    def afford_price_range(cb,App_datasets):
        bldg_name = cb['Bldg Proje']
        # Check IF WIRE!
        if bldg_name in ridt_WIRE.wire_dict.keys():
            #distrivute agents to rooms and prices
            wbd_subset = wire_bldg_data[wire_bldg_data['bld_name']==bldg_name].reset_index(drop=True) #Wire Buidling Data Subset
            f_c=0 #floor_count
            for j in wbd_subset.index:
                wbds_room_subset = wbd_subset.loc[j]
                low_t = wbds_room_subset['Low_today']
                up_t = wbds_room_subset['Upper_today']
                r_c=wbds_room_subset['Units']+f_c #roof_count
                App_datasets.loc[f_c:r_c-1,'Low_income'] = low_t
                App_datasets.loc[f_c:r_c-1,'Upper_income'] = up_t
                f_c = r_c
        else:
            App_datasets=ridt_bldgs.afford_price_range_NotWIRE(cb,App_datasets)
        return App_datasets

```

```python
class ridt_properties:
    # default constructor 
    def __init__(self,js_bldgs):
        """Recieves json building file and returns a property dataset
        js_bldgs: json_bldgs
        """
        self.json_bldgs = js_bldgs
        self.ds = self.iterate_json_bldgs()
        
    def iterate_json_bldgs(self):
        """Iterates through each json_bldg index and creates property dataset"""
        for i in (self.json_bldgs.index):
            cb = self.json_bldgs.loc[i]#current building
            App_datasets = ridt_bldgs.generate_properties_in_bldg(cb)
            App_datasets = ridt_WIRE.updateRoomsInWire(cb,App_datasets)
            App_datasets = ridt_bldgs.afford_price_range(cb,App_datasets)
            #Need to update AMI in all apartments!!!!!!!

            if (i==0): # First iteration
                property_df = App_datasets
                #total_agents = cb['Total Unit']
            else:
                frames = [property_df,App_datasets]
                property_df = pd.concat(frames)
        property_df.reset_index(inplace=True,drop=True)
        return property_df
    
    def report(self):
        check_ds_gb = self.ds.groupby('Building Name')
        check_ds = check_ds_gb.agg({'app_index':'count','built_year':'first','affordable_living':'sum',
                                    'elderly_lving':'sum','Rental_Units':'sum',
                                    'Ap_sq_feet_size':'mean',
                                    'Ap_number_of_rooms':'mean',
                                    'Low_income':'mean',
                                    'Upper_income':'mean'}).rename(
            columns={'app_index':'Apartments',
                     'built_year':'Built Year',
                     'affordable_living':' Affordable Living',
                     'elderly_lving':'Elderlies',
                     'Rental_Units':'Rentals',
                     'Ap_sq_feet_size':'Mean Sq Feet',
                     'Ap_number_of_rooms':'mean Rooms',
                     'Low_income':'Lower $',
                     'Upper_income':'Upper $'}).reset_index()
        
        for f in ['Mean Sq Feet','mean Rooms','Lower $','Upper $']:
            check_ds[f] = check_ds[f].apply(lambda x: round(x,1))

        return check_ds
```

```python
for i in (json_bldgs.index):
    #Iterate through the buildings
    cb = json_bldgs.loc[i]#current building
    total_agents = cb['Total Unit'] #Total number of Agents in the building
    bld_key = [cb['bldgs_id']]*total_agents #Defines the number of the building
    agent_dict = {'app_index' : [i for i in range(total_agents)],
                  'bld_key' : bld_key} # define the dictionary for the dataset

    App_datasets = pd.DataFrame.from_dict(agent_dict)
    #The building key and the door number
    App_datasets['address_door']=App_datasets['bld_key']+"_"+App_datasets['app_index'].astype(str)
    #I think this might be redundant
    App_datasets['Agent_Status']='Living'
    #What is the agents status in the "Action" Range
    App_datasets['agent_action']= "In Building"
    construct_year = pd.to_datetime(cb['cnstrct_yr'].astype(str)+"-1-"+"1")
    App_datasets['start_action']= construct_year
    #create a distrbution of normal ages who enter the building
    
    #ages_in_building = sample_age_according_to_ri_pattern(total_agents)
    ages_in_building = ridt_demographic.sample_ages(total_agents,'Young')
    #Birthdate is the construction - entrance age
    App_datasets['birth_date'] = [ (str(construct_year.year-age)+"-"+str(random.randint(1,12))+"-"+str(random.randint(1,28)))  for age in ages_in_building ]
    App_datasets['birth_date'] = pd.to_datetime(App_datasets['birth_date'])

    #calculates the death date based on the birthdate
    App_datasets['death_date'] = App_datasets.apply(ridt_demographic.death_date_calculator2,axis=1)
    #for statistics calculates the int actuall death age of the agent
    App_datasets['death_age'] = (((App_datasets['death_date'] - App_datasets['birth_date']).dt.days)/365).astype(int)
    #the end action by deafault is the death date (if not other incidentoccured)
    App_datasets['end_action']= App_datasets['death_date']
    #agent entrance age
    App_datasets['entrance_age'] = (((construct_year - App_datasets['birth_date']).dt.days)/365).astype(int)
    # by default all agents are not affordable living
    App_datasets['affordable_living'] = 0
    #apply random income to all agents at start
    App_datasets['income'] =  App_datasets.apply(lambda x: int(random.uniform(1.48,2.0)*ridt_info.ami_100_percent), axis=1) 
    #elederly lviing are marked
    App_datasets['elderly_lving'] = 0
    #Mark rental units
    App_datasets['Rental_Units'] = 0
    #by default appartments are not rental
    App_datasets['ApartmentType'] = 'Other'
    #Number of cycles that tenants lived in a current building
    App_datasets['Ap_tenant_cycle'] = 0
    App_datasets['Ap_sq_feet_size'] = 1364 # default
    App_datasets['Ap_number_of_rooms'] = 2 # default
    if cb['protected']=='1':
        App_datasets['protected']=True
    else:
        App_datasets['protected']=False
    
    if cb['Rental_Units']>0:
        App_datasets.loc[0:cb['Rental_Units']-1,'Rental_Units'] = 1
        App_datasets.loc[0:cb['Rental_Units']-1,'ApartmentType'] = 'Rental'


    bldg_name = cb['Bldg Proje']
    # Check IF WIRE!
    if bldg_name in ridt_WIRE.wire_dict.keys():
        print('in')
        print(bldg_name)
        print('total agents :' +str(total_agents))
        #distrivute agents to rooms and prices
        wbd_subset = wire_bldg_data[wire_bldg_data['bld_name']==bldg_name].reset_index(drop=True) #Wire Buidling Data Subset
        f_c=0 #floor_count
        for j in wbd_subset.index:
            wbds_room_subset = wbd_subset.loc[j]
            elderly_flag=0
            if (wbds_room_subset['Elderly']==1):
                elderly_flag=1
            print('Bldg: '+ wbds_room_subset['bld_name'])
            print('Bedroom : '+ str(wbds_room_subset['Bedroom']))
            print('Units : '+ str(wbds_room_subset['Units']))
            print('Elderly : '+ str(wbds_room_subset['Elderly']))
            low_t = wbds_room_subset['Low_today']
            up_t = wbds_room_subset['Upper_today']
            r_c=wbds_room_subset['Units']+f_c #roof_count
            print(str(low_t)+' '+str(up_t))
            print(str(f_c)+' '+str(r_c))
            print('----------------------')
            App_datasets.loc[f_c:r_c-1,'elderly_lving'] = elderly_flag
            App_datasets.loc[f_c:r_c-1,'income'] = App_datasets.loc[f_c:r_c-1,'income'].apply(lambda x: random.randint(int(low_t),int(up_t)))
            App_datasets.loc[f_c:r_c-1,'Ap_number_of_rooms'] =wbds_room_subset['Bedroom'] 
            App_datasets.loc[f_c:r_c-1,'Ap_sq_feet_size'] =wbds_room_subset['sq_feet'] 
            f_c = r_c
    else:
        #calculate affordability according to the building
        if cb['Afford_uni']>0:
            App_datasets.loc[0:cb['Afford_uni'],'affordable_living'] = 1
            ridt_populationDF.calculate_income_from_affordability(cb)
    
    ela = cb['elderly_affordable'] #elderly living agents
    if ela>0:
        # Check if it's wire if not all other cases
        App_datasets.loc[0:ela,'elderly_lving'] = 1
        old_ages_in_building = ridt_demographic.sample_ages(ela+1,'Old')
        con1 = App_datasets['elderly_lving']==1 #affordable living CONDITION
        App_datasets.loc[con1, 'birth_date'] = pd.to_datetime([ (str(construct_year.year-age)+"-"+str(random.randint(1,12))+"-"+str(random.randint(1,28))) for age in old_ages_in_building ])
        App_datasets.loc[con1,'death_date'] = App_datasets.loc[con1].apply(ridt_demographic.death_date_calculator2,axis=1)
        App_datasets.loc[con1,'death_age'] = (((App_datasets.loc[con1,'death_date'] - App_datasets.loc[con1,'birth_date']).dt.days)/365).astype(int)
        App_datasets.loc[con1,'end_action'] = App_datasets.loc[con1,'death_date']
        App_datasets.loc[con1,'entrance_age'] = (((construct_year - App_datasets.loc[con1,'birth_date']).dt.days)/365).astype(int)


    if (i==0): # First iteration
        df = App_datasets
        #total_agents = cb['Total Unit']
    else:
        frames = [df,App_datasets]
        df = pd.concat(frames)
df.reset_index(inplace=True,drop=True)
df['agentID'] = [uuid.uuid4() for _ in range(len(df.index))]
df['previous_agentID']='0'
df['comment']=''



```

### Simulate Agent Dynamics based on Dead AGents
### 1. Mark the dead agents as a new event in the dynaic Population (also add it to an avent diary)
### 2. List vacanat appartment
### 3. Generate new Agents who will populate the old appartment
### 4. Check if Appartment is not "elderly"


```python
#dead agent to replace
last_date= pd.to_datetime((datetime.date(2100, 1, 1)))
datr = df.loc[df['death_date']<last_date].copy() # select all the dead agents Simulation Ends
iteration = 0
new_agents_accumulator = pd.DataFrame()
#datr.to_excel('test_datr.xlsx')
```

```python
datr['death_date'].dt.year.describe()
```

```python
#set start action to be death_date + 1
# set Age of agent acccording to profile (if they are not elderly)
# set birthdate start action - age
# set death date according to birthdate
while ((iteration <15)):
    #print('First Iteration Count Total Agents: ',datr['agentID'].count())

    #change Unique ID
    if (iteration > 0):
        new_agents_accumulator = pd.concat([new_agents_accumulator,datr]) # if Iteration beyond 0 append previous list
        datr = datr.loc[datr['death_date']<last_date].copy() #select dead agents from previous iteration
    print('Count Total Agents: ',datr['agentID'].count())
    if (datr['agentID'].count()<5):
        iteration=30
        break
    datr['previous_agentID'] = datr['agentID'] #get idea of the previous agents
    datr['agentID'] = [uuid.uuid4() for _ in range(len(datr.index))] # get id of the new agents
    datr['start_action'] = datr['death_date'].apply(lambda x: x+datetime.timedelta(days=1)) #new agents move in a day after the old agents die
    #datr['start_action'] = datr['death_date']
    datr['Agent_Status']=='Living'
    datr['Ap_tenant_cycle'] +=1
    #Condition 1 Select Young Agents
    con0 = datr['elderly_lving']==0 # young agents
    pop_size = datr.loc[con0,'agentID'].count() #get count size of the 
    if pop_size > 0:
        ages = ridt_demographic.sample_ages(pop_size,'Young')
        entrance_year = pd.to_datetime(datr.loc[(con0),'start_action']).dt.year
        birth_year = entrance_year - ages
        datr.loc[con0,'birth_date'] =  [ pd.to_datetime(datetime.date(i, 1, 1)) for i in birth_year ]
        datr.loc[con0,'death_date'] =  datr.loc[(con0)].apply(ridt_demographic.death_date_calculator2,axis=1)
        datr.loc[con0,'death_date'] = datr.loc[con0,'death_date'].apply(ridt_demographic.setMaxDeathAge)
        
        datr.loc[con0,'death_age'] = (
            pd.to_datetime(datr.loc[con0,'death_date']).dt.year - 
            pd.to_datetime(datr.loc[con0,'birth_date']).dt.year)
        datr.loc[(con0),'entrance_age'] = ages
       

    # SubSet for Elderly Agents
    con1 = datr['elderly_lving']==1 # elderly living condition 1
    pop_size_1 = datr.loc[con1,'agentID'].count()
    if pop_size_1 > 0:
        ages_old = ridt_demographic.sample_ages(pop_size_1,'Old')
        entrance_year = pd.to_datetime(datr.loc[(con1),'start_action']).dt.year
        birth_year = entrance_year - ages_old
        datr.loc[con1,'birth_date'] =  [ datetime.date(i, 1, 1) for i in birth_year ]
        datr.loc[con1,'death_date'] =  datr.loc[(con1)].apply(ridt_demographic.death_date_calculator2,axis=1)
        datr.loc[con1,'death_date'] =  datr.loc[con1,'death_date'].apply(ridt_demographic.setMaxDeathAge)
        datr.loc[con1,'death_age'] = (
            pd.to_datetime(datr.loc[con1,'death_date']).dt.year - 
            pd.to_datetime(datr.loc[con1,'birth_date']).dt.year)
        datr.loc[(con1),'entrance_age'] = ages_old
    if (pop_size_1+pop_size)>0:
        datr['end_action'] = datr['death_date']
        #select the affordabile agents and update their income
        datr.loc[datr['affordable_living']==1,'income'] = datr[datr['affordable_living']==1].apply(ridt_agent.get_agents_new_income_NonWIRE, axis=1)
        datr.loc[datr['affordable_living']==0,'income'] = datr[datr['affordable_living']==0].apply(lambda x: int(random.uniform(1.48,1.8)*ridt_info.ami_100_percent), axis=1)
        iteration = iteration +1
    
new_agents_accumulator = pd.concat([new_agents_accumulator,datr])
```

## Pivoting Agents to Buildings according to the time

```python
living_agents = living_agents.copy()
dead_agents = dead_agents.copy()

living_agents['age']=(current_date_range_1-pd.to_datetime(living_agents['birth_date'])).dt.days
living_agents['age'] = living_agents['age'].apply(lambda x: int(float(x)/365))

la_groups = living_agents.groupby('bld_key') #living agents group by
d_groups = dead_agents.groupby('bld_key') #dead agents group by

la_pivot = la_groups.agg({'agentID':['count'],
              'age':['mean'],'affordable_living':['sum'],'income':['mean']})

d_pivot = d_groups.agg({'agentID':['count']})

la_pivot.columns = ['Number of Living Agents','Mean Age', 'Affordable Living','Average Income']
d_pivot.columns = ['Number of Dead Agents']

la_pivot.reset_index(inplace=True)
d_pivot.reset_index(inplace=True)
la_pivot['Affordable_ratio'] = (la_pivot['Affordable Living']/la_pivot['Number of Living Agents']).apply(lambda x:round(x,2))

agents_age = pd.merge(la_pivot,d_pivot,on='bld_key')

jbbf1 = pd.merge(json_bldgs,agents_age,left_on='bldgs_id', right_on='bld_key',how='right')

jbbf1['Mean Age'] = jbbf1['Mean Age'].astype(int)
```
