```python
def random_date(start_date,end_date):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date_ = start_date + datetime.timedelta(days=random_number_of_days)

    return random_date_
```

```python
def death_date(birth_date):
    #start_date = datetime.date(1920, 1, 1)
    #end_date = datetime.date(1950, 1, 1)
    random_number_of_days = random.randrange(60*365,stop=100*365)
    death_date_ = birth_date + datetime.timedelta(days=random_number_of_days)

    return death_date_
```

```python
def death_date_calculator(bd,a = 0.4,b = 48):
    mdy = int((bd.year-1900) * a + b) # mdy manhattan death year
    death_date_ = datetime.date(mdy+bd.year, 1, 1)
    random_number_of_days = np.random.normal(scale=1)*20*365 #+- 20 years
    death_date_ = death_date_ + datetime.timedelta(days=random_number_of_days)
    if (death_date_.year)<1975:
        death_date_ =  death_date_ + datetime.timedelta(days=(1975-death_date_.year)*365)
    return death_date_
    
    
```

```python
def death_date_calculator2(ag,a = 0.4,b = 48):
    bd = ag.birth_date
    ed = ag['start_action']
    
    mdy = int((bd.year-1900) * a + b) # mdy manhattan death year
    death_date_ = datetime.date(bd.year+mdy, 1, 1)
    random_number_of_days = np.random.normal(scale=1)*20*365 #+- 20 years
    death_date_ = death_date_ + datetime.timedelta(days=random_number_of_days)
    if (death_date_)<ed:
        death_date_ =  death_date_ + datetime.timedelta(days=(ed.year-death_date_.year+random.randrange(1,5))*365)
    return death_date_
    
```

```python
def get_group_category(x):
    if (x>85):
        return 90
    elif (x>75):
        return 80
    elif (x>65):
        return 70
    elif (x>60):
        return 62
    elif (x>55):
        return 57
    elif (x>45):
        return 50
    elif(x>35):
        return 40
    elif(x>25):
        return 30
    elif(x>20):
        return 22
```

```python
def age_group_fun(num):
    return age_group_var[num]
```

```python
def sample_elderly_age(pop_size):
    sam = np.random.choice(age_group_elderly, pop_size, p=p_elderly)
    sam_variance = np.array(list(map(age_group_fun, sam)))*np.random.normal(scale=0.37,size=pop_size)
    sam2 = sam + sam_variance
    return sam2.astype(int)
```

```python
def sample_age_according_to_ri_pattern(pop_size):
    sam = np.random.choice(age_group, pop_size, p=p)
    sam_variance = np.array(list(map(age_group_fun, sam)))*np.random.normal(scale=0.37,size=pop_size)
    sam2 = sam + sam_variance
    return sam2.astype(int)
```

```python
def sample_age_according_to_ri_pattern2(pop_size, ag = age_group,p_ = p):
    sam = np.random.choice(ag, pop_size, p=p_)
    sam_variance = np.array(list(map(age_group_fun, sam)))*np.random.normal(scale=0.37,size=pop_size)
    sam2 = sam + sam_variance
    return sam2.astype(int)
```

```python
def birth_date(entrance_date):
    #start_date = datetime.date(1920, 1, 1)
    #end_date = datetime.date(1950, 1, 1)
    random_number_of_days = random.randrange(25*365,stop=65*365)
    birth_date_ = entrance_date - datetime.timedelta(days=random_number_of_days)

    return birth_date_
```

# Build Synthetic Agents

```python
for i in (json_bldgs.index):
    cb = json_bldgs.loc[i]#current building
    total_agents = cb['Total Unit']
    bld_key = [cb['bldgs_id']]*total_agents

    agent_dict = {'app_index' : [i for i in range(total_agents)],
                  'bld_key' : bld_key} # define the dictionary for the dataset

    App_datasets = pd.DataFrame.from_dict(agent_dict)
    App_datasets['address_door']=App_datasets['bld_key']+"_"+App_datasets['app_index'].astype(str)
    App_datasets['Agent_Status']='Living'
    App_datasets['agent_action']= "In Building"
    construct_year = datetime.date(cb['cnstrct_yr'], 1, 1)
    App_datasets['start_action']= construct_year
    
    #App_datasets['birth_date'] = App_datasets['start_action'].apply(birth_date)
    ages_in_building = sample_age_according_to_ri_pattern(total_agents)
    App_datasets['birth_date'] = [ construct_year - datetime.timedelta(days=int(age)*365) for age in ages_in_building ]
    App_datasets['death_date'] = App_datasets.apply(death_date_calculator2,axis=1)
    App_datasets['death_age'] = (((App_datasets['death_date'] - App_datasets['birth_date']).dt.days)/365).astype(int)
    App_datasets['end_action']= App_datasets['death_date']
    App_datasets['entrance_age'] = (((construct_year - App_datasets['birth_date']).dt.days)/365).astype(int)
    App_datasets['affordable_living'] = 0
    App_datasets['elderly_lving'] = 0
    App_datasets['Rental_Units'] = 0
    App_datasets['ApartmentType'] = 'Other'
    if cb['Rental_Units']>0:
        App_datasets.loc[0:cb['Rental_Units']-1,'Rental_Units'] = 1
        App_datasets.loc[0:cb['Rental_Units']-1,'ApartmentType'] = 'Rental'
        
    if cb['Afford_uni']>0:
        App_datasets.loc[0:cb['Afford_uni'],'affordable_living'] = 1
    ela = cb['elderly_affordable'] #elderly living agents
    if ela>0:
        App_datasets.loc[0:ela,'elderly_lving'] = 1
        old_ages_in_building = sample_age_according_to_ri_pattern2(ela+1,age_group_elderly,p_elderly)
        con1 = App_datasets['elderly_lving']==1 #affordable living CONDITION
        App_datasets.loc[con1, 'birth_date'] = [ construct_year - datetime.timedelta(days=int(age)*365) for age in old_ages_in_building ]
        App_datasets.loc[con1,'death_date'] = App_datasets.loc[con1].apply(death_date_calculator2,axis=1)
        App_datasets.loc[con1,'death_age'] = (((App_datasets.loc[con1,'death_date'] - App_datasets.loc[con1,'birth_date']).dt.days)/365).astype(int)
        App_datasets.loc[con1,'entrance_age'] = (((construct_year - App_datasets.loc[con1,'birth_date']).dt.days)/365).astype(int)
    if (cb['apply_income']==True):
        App_datasets['income'] = cb['affordable_max']
        App_datasets['income'] = App_datasets['income'].apply(lambda x: x + random.randint(0, 50000))
        ed = cb['Afford_uni'].astype(int)
        #ed = int(total_agents* cb['afford_percent']/100.0) #eligable for discount
        low_income =  [random.randint(cb['affordable_min'],cb['affordable_max']) for i in range(0,ed)]
        App_datasets.loc[0:ed-1,'income'] = low_income
        #App_datasets.loc[0:ed-1,'ApartmentType'] = affordable_category[cb['Category']]
        #App_datasets.loc[ed:,'ApartmentType'] = Non_affordable_category[cb['Category']]
    #App_dataset[Renter,owner/Condo,Coop]
    #App_dataset[income]
    #Wire were ha
    if (i==0): # First iteration
        df = App_datasets
        #total_agents = cb['Total Unit']
    else:
        frames = [df,App_datasets]
        df = pd.concat(frames)
df.reset_index(inplace=True,drop=True)
df['agentID'] = [uuid.uuid4() for _ in range(len(df.index))]
df['previous_agentID']='0'



```

```python
datr = df.loc[df['death_date']<end_of_simulation_date].copy() #dead agent to replace
iteration = 0
new_agents_accumulator = pd.DataFrame()
```

```python
#set start action to be death_date + 1
# set Age of agent acccording to profile (if they are not elderly)
# set birthdate start action - age
# set death date according to birthdate
while ((iteration <8)):
    #change Unique ID
    if (iteration > 0):
        new_agents_accumulator = pd.concat([new_agents_accumulator,datr])
        datr = datr.loc[datr['death_date']<end_of_simulation_date].copy() #dead agent to replace
    datr['previous_agentID'] = datr['agentID']
    datr['agentID'] = [uuid.uuid4() for _ in range(len(datr.index))]
    datr['start_action'] = datr['death_date'].apply(lambda x: x+datetime.timedelta(days=1))
    con0 = datr['elderly_lving']==0 # young agents
    pop_size = datr.loc[con0,'agentID'].count()
    if pop_size > 0:
        ages = sample_age_according_to_ri_pattern2(pop_size,age_group,p)
        entrance_year = pd.to_datetime(datr.loc[(con0),'start_action']).dt.year
        birth_year = entrance_year - ages
        datr.loc[con0,'birth_date'] =  [ datetime.date(i, 1, 1) for i in birth_year ]
        datr.loc[con0,'death_date'] =  datr.loc[(con0)].apply(death_date_calculator2,axis=1)
        datr.loc[con0,'death_age'] = (
            pd.to_datetime(datr.loc[con0,'death_date']).dt.year - 
            pd.to_datetime(datr.loc[con0,'birth_date']).dt.year)
        datr.loc[(con0),'entrance_age'] = ages

    # SubSet for Elderly Agents
    con1 = datr['elderly_lving']==1 # elderly living condition 1
    pop_size_1 = datr.loc[con1,'agentID'].count()
    if pop_size_1 > 0:
        ages = sample_age_according_to_ri_pattern2(pop_size_1,age_group_elderly,p_elderly)
        entrance_year = pd.to_datetime(datr.loc[(con1),'start_action']).dt.year
        birth_year = entrance_year - ages
        datr.loc[con1,'birth_date'] =  [ datetime.date(i, 1, 1) for i in birth_year ]
        datr.loc[con1,'death_date'] =  datr.loc[(con1)].apply(death_date_calculator2,axis=1)
        datr.loc[con1,'death_age'] = (
            pd.to_datetime(datr.loc[con1,'death_date']).dt.year - 
            pd.to_datetime(datr.loc[con1,'birth_date']).dt.year)
        datr.loc[(con1),'entrance_age'] = ages
    datr['end_action'] = datr['death_date']
    iteration = iteration +1
```

## Pivoting Agents to Buildings according to the time

```python
living_agents = living_agents.copy()
dead_agents = dead_agents.copy()

living_agents['age']=(current_date_range_1-living_agents['birth_date']).dt.days
living_agents['age'] = living_agents['age'].apply(lambda x: int(float(x)/365))

la_groups = living_agents.groupby('bld_key') #living agents group by
d_groups = dead_agents.groupby('bld_key') #dead agents group by

la_pivot = la_groups.agg({'agentID':['count'],
              'age':['mean'],'affordable_living':['sum'],'income':['mean']})

d_pivot = d_groups.agg({'agentID':['count']})

la_pivot.columns = ['Number of Living Agents','Mean Age', 'Affordable Living','Average Income']
d_pivot.columns = ['Number of Dead Agents']

la_pivot.reset_index(inplace=True)
d_pivot.reset_index(inplace=True)
la_pivot['Affordable_ratio'] = (la_pivot['Affordable Living']/la_pivot['Number of Living Agents']).apply(lambda x:round(x,2))

agents_age = pd.merge(la_pivot,d_pivot,on='bld_key')

jbbf1 = pd.merge(json_bldgs,agents_age,left_on='bldgs_id', right_on='bld_key',how='right')

jbbf1['Mean Age'] = jbbf1['Mean Age'].astype(int)
```
